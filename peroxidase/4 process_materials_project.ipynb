```python
import pandas as pd
from mp_api.client import MPRester
```

```python
mp = pd.read_excel('./data/materials_project.xlsx')
```

```python
api_key = 'Your API'

docs = []
with MPRester(api_key) as mpr:
    for i in mp.mp_id.tolist():
        res = mpr.summary.search(material_ids=i,
                                 fields=['material_id', 'formula_pretty', 'theoretical', 
                                         'symmetry', 'volume', 'density',
                                         'band_gap', 'efermi', 'formation_energy_per_atom'])
        docs.append(res)
```

```python
rows = []

for d in docs:
    rows.append({
        "formula": d[0].formula_pretty,
        "pearson": d[0].symmetry.crystal_system.value,
        "space": d[0].symmetry.number,
        "density": d[0].density,
        "volume": d[0].volume,
        "band_gap": d[0].band_gap,
        "efermi": d[0].efermi,
        "formation_energy_per_atom": d[0].formation_energy_per_atom,
    })
```

```python
mp_features = pd.DataFrame(rows)
```

```python
mp_features.to_csv('./data/materials_project_features.csv')
```
